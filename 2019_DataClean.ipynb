```python
import pandas as pd
import os, glob
```

```python
# Merging csv files (without repeating header)
data_files = glob.glob("data/*.csv") 
df = pd.concat((pd.read_csv(f, header = 0) for f in data_files), ignore_index= True)
df
```

```python
# Export file as a CSV
df.to_csv("data/2019-citibike-tripdata.csv")
```
